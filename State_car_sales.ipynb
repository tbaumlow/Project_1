```python
# Create Dataframe from the car sales price data

import pandas as pd
car_df = pd.read_csv('/car_prices.csv')
car_df
```

```python
# Describe the numeric columns in the data set
car_df.describe()
```

```python
# Count the unique vehicle makes

car_df['state'].nunique()
```

```python
# Top 10 States by Cars sold

import matplotlib.pyplot as plt

# Create a bar chart of the top 10 states by count
car_df['state'].value_counts().head(10).plot(kind='bar')

# Add a title and labels
plt.title('Top 10 States by Car Sales')
plt.xlabel('State')
plt.ylabel('Count')

# Rotate the x-axis labels for readability
plt.xticks(rotation=45)

# Display the plot
plt.show()
```

```python
# prompt: show the top 10 states by average sellingprice

# Calculate the average selling price for each state
average_selling_price_by_state = car_df.groupby('state')['sellingprice'].mean()

# Sort the states by average selling price in descending order
top_10_states_by_avg_selling_price = average_selling_price_by_state.sort_values(ascending=False).head(10)

# Create a bar chart of the top 10 states by average selling price
top_10_states_by_avg_selling_price.plot(kind='bar')

# Add a title and labels
plt.title('Top 10 States by Average Selling Price')
plt.xlabel('State')
plt.ylabel('Average Selling Price')

# Rotate the x-axis labels for readability
plt.xticks(rotation=45)

# Display the plot
plt.show()
```
