```python
# Create Dataframe from the car sales price data

import pandas as pd
car_df = pd.read_csv('car_prices.csv')
car_df
```

```python
# Describe the numeric columns in the data set
car_df.describe()
```

```python
# Count the unique vehicle makes

car_df['make'].nunique()
```

```python
# graph the top 10 vehicle makes by count with a bar graph

car_df['make'].value_counts().head(10).plot(kind='bar')
```

```python
# The top 10 vehicle makes by odometer

car_df.groupby('make')['odometer'].mean().sort_values(ascending=False).head(10).plot(kind='bar')
```

```python
# The top 10 vehicle makes by average selling price

average_selling_price_by_make = car_df.groupby('make')['sellingprice'].mean().sort_values(ascending=False)

average_selling_price_by_make.head(10).plot(kind='bar')
```

```python
# Scatter plot to show the odometer vs selling price

car_df.plot.scatter(x='sellingprice', y='odometer')
```
